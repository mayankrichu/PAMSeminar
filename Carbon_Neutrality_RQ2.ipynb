```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV data into a pandas DataFrame
data = pd.read_csv('/AIR_GHG_07072023190521703.csv')

# Extract the years and carbon release percentage columns from the DataFrame
years = data['Year'].values
carbon_release = data['Value'].values

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(years, carbon_release)

# Calculate the predicted values using the regression line
predicted_years = np.arange(2021, 2051)
predicted_release = slope * predicted_years + intercept

# Plotting the data and regression line
plt.plot(years, carbon_release, 'bo', label='Carbon Release Percentages')
plt.plot(predicted_years, predicted_release, 'r-', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Carbon Release Percentage')
plt.title('Carbon Release Percentages and Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

print(f"Slope of the regression line: {slope}")

# Printing the predicted values for years 2021-2050
print("Predicted values for years 2021-2050:")
for year, release in zip(predicted_years, predicted_release):
    print(f"Year: {year}, Predicted Release: {release}")
```
